# 1. Funções

As funções são o principal e mais importante método de organização de código e reutilização em Python;

Como regra geral, se você prevê que precisará repetir o mesmo código ou um código semelhante mais de uma vez, talvez valha a pena escrever uma função reutilizável;

Função é um dispositivo que agrupa um conjunto de instruções para que elas possam ser executadas mais de uma vez;

Funções também permitem especificar os parâmetros que podem servir como entrada para as funções.

Definindo uma função em Python:

```python
def my_function(x, y, z=1.5):
  if z < 1:
    return z * (x + y)
  else:
    return z / (x + y)
```

````python
# soma
def num_Soma(x,y):
  return x+y


print( num_Soma(1, 5) )
````

````python
# série de Fibonacci até um limite arbitrário 'n'
# n = 50: 0 1 1 2 3 5 8

def fib(x):
  a = 0
  b = 1
  lista = []
  while a < x:
    #print(a, end=' ')
    lista.append(a)
    temp = a
    a = b
    b = b + temp
  #print()
  return lista



l = fib(1000)
print(type(l))

````
